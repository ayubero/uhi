```python
import pandas as pd
import folium
from pyproj import Transformer
import os
```

```python
# Folder containing the raw station files
folder_path = 'raw_stations'

# Get the list of CSV files (without the .csv extension)
existing_files = [f.replace('.csv', '') for f in os.listdir(folder_path) if f.endswith('.csv')]
print('Number of stations', len(existing_files))
```

```python
stations = pd.read_csv('netatmo_stations.csv', delimiter=',')
stations.head()
```

```python
# Filter stations to keep only rows where device_id exists as a file
stations = stations[stations['device_id'].isin(existing_files)]
len(stations)
```

```python
madrid_coords = (40.4168, -3.7038)
map = folium.Map(location=madrid_coords, zoom_start=13)

# Iterate over each station
for index, station in stations.iterrows():
    lon = station['lon']
    lat = station['lat']

    # Add a marker for the point
    folium.Marker([lat, lon], popup=station['device_id']).add_to(map)

# If running in a Jupyter Notebook, display the map inline
map
```

### Generate shapefile

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(stations['lon'], stations['lat'])]
gdf = gpd.GeoDataFrame(stations, geometry=geometry)

# Set CRS and save
filename = 'netatmo_stations.shp'
gdf.set_crs(epsg=4326, inplace=True)
gdf = gdf.to_crs('EPSG:25830')
gdf.to_file(filename)
```
