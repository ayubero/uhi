```python
import matplotlib.pyplot as plt
import rasterio
```

## Zaragoza

```python
raster_path = '../data/zaragoza/results/universal_kriging_without_projection.tif'

with rasterio.open(raster_path) as src:
    no_data_value = src.nodata
    band = src.read(1)

plt.figure(figsize=(10, 10))
plt.imshow(band, cmap='plasma')
plt.colorbar(label='º C')
plt.show()
```
