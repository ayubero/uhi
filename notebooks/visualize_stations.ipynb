```python
import pandas as pd
import folium
from pyproj import Transformer
```

```python
city = 'orleans'
stations = pd.read_csv('../data/' + city + '/stations/netatmo_stations.csv', delimiter=',')
stations.head()
```

```python
stations['lat'].mean()
```

```python
coords = (stations['lat'].mean(), stations['lon'].mean())
map = folium.Map(location=coords, zoom_start=13)

# Iterate over each station
for index, station in stations.iterrows():
    lon = station['lon']
    lat = station['lat']

    # Add a marker for the point
    folium.Marker([lat, lon], popup=station['device_id']).add_to(map)

# If running in a Jupyter Notebook, display the map inline
map
```

### Generate shapefile

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(stations['lon'], stations['lat'])]
gdf = gpd.GeoDataFrame(stations, geometry=geometry)

# Set CRS and save
filename = '../data/' + city + '/shapefiles/netatmo_stations.shp'
gdf.set_crs(epsg=4326, inplace=True)
gdf = gdf.to_crs('EPSG:4326')
gdf.to_file(filename)
```
